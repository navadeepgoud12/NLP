#Tokenisation is a process to convert either a paragraph or sentences into tokens. means if we convert paragraph into sentences we can also convert paragraph into words if we convert into single single word becomes tokens if sentences then every sentence is token

```python
import nltk
nltk.download('punkt')
```

```python
corpus = """Hello welcome, iam navadeep goud. now i started NLP in my datascincejourney!
iam very excited!.
"""
```

```python
print(corpus)
```

```python
## tokenisation 
## sentence-->paragraph
#sent_tokenize is the package that convert sentence into paragraph
from nltk.tokenize import sent_tokenize

```

```python
nltk.download('punkt_tab')
```

```python
documents = sent_tokenize(corpus)
```

```python
print(type(documents))
```

```python
print(documents)
```

```python
for doc in documents:
    print(doc)
```

```python
# tokenisation
## paragraph into words
##sentence ---> words
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for doc in documents:
    print(word_tokenize(doc))
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
