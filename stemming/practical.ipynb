Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of the words known as lemma

```python
## comments of product is a positive or negitive review
# reviews -----> eat,eating,eats,go,goes,going
```

## PorterStemer

```python
words = ['eating','eat','eats','drink','drinking','program','programs']
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
for word in words:
    print(word+"------>"+stemmer.stem(word))
```

```python
stemmer.stem('congratulations')
```

```python
stemmer.stem("helpful")
```

```python
## RegexStemmer

from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|es$|able',min=4)
```

```python
reg_stemmer.stem("navas")
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("navable")
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowballStemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowballStemmer.stem(word))
```

```python
stemmer.stem('fairly'),stemmer.stem('sportingly')
```

```python
snowballStemmer.stem('fairly'),snowballStemmer.stem('sportingly')
```

WordNet Lemmatizer is an NLP technique that reduces words to their base or dictionary form (lemma) using lexical knowledge from the WordNet database.

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize("history",pos='v')
```
